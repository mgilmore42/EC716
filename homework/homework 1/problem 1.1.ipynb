```python
from scipy.io import wavfile

import matplotlib.pyplot as plt

import numpy as np
```

```python
fs, data_2channel = wavfile.read('audio/Problem 1.wav')

data = data_2channel[:,0]
```

```python
plt.plot(np.arange(len(data))/fs, data/data.max())

plt.title('Saying Prompt Sampled at 16kHz')

plt.xlabel('time (s)')
plt.ylabel('normalized amplitude')

plt.savefig('plots/Problem 1.svg')

plt.show()
```
